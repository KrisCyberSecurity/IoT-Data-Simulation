```python
import random
import time
import pandas as pd

def generate_iot_data(samples=10):
    data = []

    for i in range(samples):
        device_id = f"Device_{i+1}"
        temperature = round(random.uniform(20, 35), 2)
        humidity = random.randint(40, 80)
        air_quality = random.randint(50, 200)
        status = random.choice(["Normal", "Warning", "Critical"])
        timestamp = time.strftime('%Y-%m-%d %H:%M:%S')

        data.append([device_id, temperature, humidity, air_quality, status, timestamp])

        time.sleep(1)

    df = pd.DataFrame(data, columns=[
        "Device ID",
        "Temperature",
        "Humidity",
        "Air Quality",
        "Status",
        "Timestamp"
    ])

    return df


# 👇 STEP 1: Generate data
iot_data = generate_iot_data(10)

# 👇 STEP 2: Display (optional)
iot_data

# 👇 STEP 3: SAVE TO CSV (THIS IS WHERE YOU ADD IT)
iot_data.to_csv("iot_data.csv", index=False)
```
